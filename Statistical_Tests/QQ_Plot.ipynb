# Q-Q Plot

Q-Q Plot is used to visually assess whether data follows a normal distribution.

Points close to the line indicate normality.

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

```python
scores=[72,75,78,80,70,74,77,76,79,73]
```

```python
plt.figure(figsize=(8,6))

stats.probplot(
    scores,
    dist='norm',
    plot=plt
)

plt.title(
    "Q-Q Plot",
    fontsize=16,
    fontweight='bold'
)

plt.grid()

plt.show()
```

# Interpretation

If points lie close to the reference line:

Data is approximately normally distributed.

If points deviate strongly from the line:

Data may not be normally distributed.

# Conclusion

Q-Q plots provide a visual method for checking normality assumptions.

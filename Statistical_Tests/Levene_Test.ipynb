# Levene Test

The Levene test checks whether the variances of different groups are equal.

H₀: Variances are equal.

H₁: Variances are not equal.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import levene

sns.set_style("whitegrid")
```

```python
group1=[70,72,75,78,80]

group2=[68,71,73,77,79]

group3=[65,69,74,76,81]
```

```python
statistic,p_value=levene(
    group1,
    group2,
    group3
)

print("Levene statistic =",round(statistic,3))
print("P-value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=[group1,group2,group3]
)

plt.xticks(
    [0,1,2],
    ['Group1','Group2','Group3']
)

plt.title(
    "Levene Test - Group Variances",
    fontsize=16,
    fontweight='bold'
)

plt.show()
```

# Conclusion

If p-value > 0.05

Equal variances assumption holds.

If p-value < 0.05

Variances are significantly different.

# Chi-Square Test

Used to compare observed frequencies with expected frequencies.

```python
import numpy as np
from scipy.stats import chisquare

observed=[25,35,40]

expected=[100/3,100/3,100/3]
```

```python
chi_stat,p_value=chisquare(
    observed,
    expected
)

print("Chi-square statistic =",round(chi_stat,3))

print("P-value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
labels=['A','B','C']

plt.figure(figsize=(8,5))

plt.bar(
    labels,
    observed,
    color='skyblue'
)

plt.axhline(
    y=expected[0],
    color='red',
    linestyle='--',
    linewidth=3,
    label='Expected Frequency'
)

plt.title(
    "Observed vs Expected Frequencies",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Categories")
plt.ylabel("Frequency")

plt.legend()

plt.show()
```

# Conclusion

Chi-square test checks whether observed frequencies differ significantly from expected frequencies.

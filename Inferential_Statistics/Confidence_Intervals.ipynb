# Confidence Intervals

## Introduction

A confidence interval provides a range of plausible values for a population parameter based on sample data.

Instead of estimating a single value, confidence intervals provide an interval within which the true population parameter is likely to lie.

Confidence intervals are widely used in:

- Medical research
- Quality control
- Survey analysis
- Machine learning
- Business analytics

# Objective

Estimate the population mean using sample information.

Determine the uncertainty associated with the estimate.

Construct:

- 90% Confidence Interval
- 95% Confidence Interval
- 99% Confidence Interval

# Formula

Confidence Interval

CI = x̄ ± z(σ/√n)

where

x̄ = Sample Mean

σ = Population Standard Deviation

n = Sample Size

z = Critical Value

# Common z-values

| Confidence Level | z-value |
|-----------------|---------|
| 90% | 1.645 |
| 95% | 1.96 |
| 99% | 2.576 |

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("talk")

plt.rcParams['figure.figsize']=(10,6)
```

# Sample Data

Suppose a sample of exam scores has:

Sample Mean = 75

Population Standard Deviation = 10

Sample Size = 100

```python
sample_mean = 75

sigma = 10

n = 100
```

# Compute Margin of Error

Margin of Error

ME = z × (σ/√n)

```python
z = 1.96

margin_error = z*(sigma/np.sqrt(n))

print("Margin of Error =",round(margin_error,3))
```

# Compute Confidence Interval

```python
lower = sample_mean-margin_error

upper = sample_mean+margin_error

print("Lower Bound =",round(lower,2))

print("Upper Bound =",round(upper,2))
```

# Interpretation

We are 95% confident that the true population mean lies between 73.04 and 76.96.

```python
plt.figure(figsize=(10,3))

plt.errorbar(
    sample_mean,
    1,
    xerr=margin_error,
    fmt='o',
    capsize=12,
    markersize=12,
    linewidth=3,
    color='royalblue'
)

plt.title(
    "95% Confidence Interval",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Population Mean")

plt.yticks([])

plt.show()
```

# Effect of Sample Size

As sample size increases:

- Standard error decreases.
- Margin of error decreases.
- Confidence interval becomes narrower.

```python
sample_sizes=[25,50,100,200]

interval_width=[]

for n in sample_sizes:

    me=1.96*(sigma/np.sqrt(n))

    interval_width.append(2*me)

interval_width
```

```python
plt.figure(figsize=(8,5))

sns.lineplot(
    x=sample_sizes,
    y=interval_width,
    marker='o',
    linewidth=3,
    color='red'
)

plt.title(
    "Effect of Sample Size on Confidence Interval Width",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Sample Size")

plt.ylabel("Interval Width")

plt.show()
```

# Effect of Confidence Level

Higher confidence level

↓

Larger z-value

↓

Larger margin of error

↓

Wider confidence interval

```python
levels=['90%','95%','99%']

widths=[3.29,3.92,5.15]

plt.figure(figsize=(8,5))

sns.barplot(
    x=levels,
    y=widths,
    palette='viridis'
)

plt.title(
    "Effect of Confidence Level on Interval Width",
    fontsize=18,
    fontweight='bold'
)

plt.ylabel("Width")

plt.show()
```

# Hypothesis Testing

## Introduction

Hypothesis testing is a statistical method used to make decisions about population parameters based on sample data.

It helps determine whether there is sufficient evidence to reject a null hypothesis.

### Applications

- Drug effectiveness
- Exam score comparison
- Quality control
- Machine learning model evaluation
- Business decision making

# Steps in Hypothesis Testing

1. State the Null Hypothesis (H₀)
2. State the Alternative Hypothesis (H₁)
3. Select significance level α
4. Compute test statistic
5. Calculate p-value
6. Make decision
7. Interpret results

# Null and Alternative Hypotheses

### Null Hypothesis H₀

There is no effect or no difference.

Example:

μ = 70

Population mean equals 70.

---

### Alternative Hypothesis H₁

There is an effect or difference.

Example:

μ ≠ 70

# Significance Level

Common significance levels:

α = 0.05

α = 0.01

Decision Rule:

If p-value < α

Reject H₀

Otherwise

Fail to reject H₀

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import ttest_1samp
```

# Sample Dataset

Exam scores from 10 students

```python
scores = [72,75,78,80,70,74,77,76,79,73]

df = pd.DataFrame({
    'Exam_Score':scores
})

df
```

```python
df.describe()
```

# Visualizing Data

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Exam_Score'],
    bins=8,
    kde=True
)

plt.title("Distribution of Exam Scores")

plt.show()
```

# One-Sample t-Test

Suppose the population mean is 70.

We want to determine whether student scores differ significantly from 70.

### H₀

μ = 70

### H₁

μ ≠ 70

```python
t_statistic,p_value = ttest_1samp(scores,70)

print("t statistic =",round(t_statistic,3))
print("p value =",round(p_value,5))
```

# Decision Rule

α = 0.05

```python
alpha = 0.05

if p_value < alpha:
    
    print("Reject H0")
    
else:
    
    print("Fail to Reject H0")
```

# Interpretation

If p-value < 0.05

Reject H₀

There is significant evidence against the null hypothesis.

If p-value > 0.05

Fail to reject H₀

There is insufficient evidence against the null hypothesis.

# Type I Error

Rejecting H₀ when H₀ is actually true.

False Positive.

Probability = α

# One-Tailed and Two-Tailed Tests

### Two-Tailed

H₁ : μ ≠ 70

Checks for differences in both directions.

---

### Right-Tailed

H₁ : μ > 70

Checks whether mean is greater.

---

### Left-Tailed

H₁ : μ < 70

Checks whether mean is smaller.

```python
mean_score = np.mean(scores)

plt.figure(figsize=(8,5))

plt.axvline(
    mean_score,
    color='red',
    label='Sample Mean'
)

sns.histplot(
    scores,
    kde=True
)

plt.legend()

plt.title("Sample Mean Visualization")

plt.show()
```

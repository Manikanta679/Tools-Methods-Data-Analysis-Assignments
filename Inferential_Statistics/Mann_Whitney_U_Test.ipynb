# Mann-Whitney U Test

The Mann-Whitney U test is a non-parametric alternative to the independent t-test.

Used when:

- Two independent groups exist.
- Data is not normally distributed.
- Comparing medians instead of means.

# Example

Compare exam scores of two classes.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import mannwhitneyu

sns.set_style("whitegrid")
```

```python
group1=[72,75,78,80,74]

group2=[65,68,70,72,67]
```

```python
u_statistic,p_value=mannwhitneyu(
    group1,
    group2
)

print("U statistic =",u_statistic)
print("P-value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=[group1,group2]
)

plt.xticks(
    [0,1],
    ['Class A','Class B']
)

plt.title("Comparison of Two Groups")

plt.show()
```

# Conclusion

If p-value < 0.05

Reject H₀.

Otherwise,

Fail to reject H₀.

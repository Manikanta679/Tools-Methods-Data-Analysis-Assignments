# Friedman Test

Non-parametric alternative to repeated measures ANOVA.

Used for comparing repeated measurements.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

sns.set_style("whitegrid")
sns.set_context("talk")

plt.rcParams['figure.figsize'] = (10,6)
```

```python
from scipy.stats import friedmanchisquare

time1=[70,72,75,78]

time2=[74,76,79,82]

time3=[78,80,84,86]
```

```python
statistic,p_value=friedmanchisquare(
    time1,
    time2,
    time3
)

print("Statistic =",round(statistic,3))

print("P-value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    time1,
    marker='o',
    label='Time1'
)

plt.plot(
    time2,
    marker='o',
    label='Time2'
)

plt.plot(
    time3,
    marker='o',
    label='Time3'
)

plt.legend()

plt.title("Repeated Measurements")

plt.show()
```

# Conclusion

Friedman test compares repeated measurements when data is non-normal.

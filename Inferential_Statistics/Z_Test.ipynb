# Z-Test

The Z-test is used to compare a sample mean with a population mean when:

- Population standard deviation is known.
- Sample size is large (n ≥ 30).

# Formula

Z = (x̄ - μ)/(σ/√n)

where

x̄ = Sample mean

μ = Population mean

σ = Population standard deviation

n = Sample size

```python
import numpy as np
from scipy.stats import norm

sample_mean = 78
population_mean = 75
sigma = 10
n = 100
```

```python
z = (sample_mean-population_mean)/(sigma/np.sqrt(n))

print("Z statistic =",round(z,3))
```

```python
p_value = 2*(1-norm.cdf(abs(z)))

print("P-value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x=np.linspace(-4,4,1000)

plt.figure(figsize=(8,5))

plt.plot(x,norm.pdf(x))

plt.axvline(
    z,
    color='red',
    linewidth=3,
    label='Z statistic'
)

plt.title("Standard Normal Distribution")

plt.legend()

plt.show()
```

# Conclusion

If p-value < 0.05

Reject H₀.

Otherwise,

Fail to reject H₀.

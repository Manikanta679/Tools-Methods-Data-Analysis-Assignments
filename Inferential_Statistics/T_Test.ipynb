# T-Test

Used when:

- Population standard deviation is unknown.
- Sample size is small.

```python
import numpy as np
from scipy.stats import ttest_1samp

scores=[72,75,78,80,70,74,77,76]
```

```python
t_statistic,p_value=ttest_1samp(scores,70)

print("t statistic =",round(t_statistic,3))

print("p value =",round(p_value,5))
```

```python
alpha=0.05

if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

# Conclusion

T-test is suitable when the population standard deviation is unknown and sample size is small.

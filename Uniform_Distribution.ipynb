# Uniform Distribution

Definition

A Uniform Distribution is a probability distribution where all outcomes in a fixed interval are equally likely.

For a continuous uniform distribution:

X∼U(a,b)

This means the random variable X can take any value between a and b, and every value has the same chance of occurring.

Example:

X∼U(0,100)

This means values from 0 to 100 are equally likely.

Basic Properties

For a continuous uniform distribution:

X∼U(a,b)
Probability Density Function — PDF
f(x)=

Outside the interval, the probability density is 0.

That means the PDF is a flat horizontal line.

Why This Dataset Is Good

This dataset is good for a uniform distribution because:

The values are generated between a fixed range, 0 to 100.

Every value in that range has an equal chance of appearing.

There is no central peak like in normal distribution.

The histogram should look roughly flat when the sample size is large.

This makes it ideal for demonstrating the basic idea of a uniform distribution.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Set random seed so results are reproducible
np.random.seed(42)

# Define the range of the uniform distribution
a = 0
b = 100

# Generate uniform samples
n = 1000
uniform_data = np.random.uniform(a, b, n)

# Print first few values
print("First 10 generated values:")
print(uniform_data[:10])

# Calculate sample mean and variance
sample_mean = np.mean(uniform_data)
sample_variance = np.var(uniform_data)

# Theoretical mean and variance
theoretical_mean = (a + b) / 2
theoretical_variance = ((b - a) ** 2) / 12

print("\nSample Mean:", sample_mean)
print("Theoretical Mean:", theoretical_mean)

print("\nSample Variance:", sample_variance)
print("Theoretical Variance:", theoretical_variance)
```

```python
uniform_data[:10]
```

```python
# Plot histogram of generated data
plt.figure(figsize=(9, 5))

plt.hist(
    uniform_data,
    bins=20,
    density=True,
    edgecolor="black",
    alpha=0.7,
    label="Generated Uniform Samples"
)

# Plot theoretical PDF line
x = np.linspace(a, b, 100)
pdf = np.ones_like(x) * (1 / (b - a))

plt.plot(
    x,
    pdf,
    linewidth=2,
    label="Theoretical Uniform PDF"
)

plt.title("Uniform Distribution: Generated Samples and Theoretical PDF")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

Explanation of the Graph

The histogram shows the generated uniform samples.

The theoretical PDF line is flat because every value between 0 and 100 has the same probability density.

The bars may not be perfectly equal because the dataset is randomly generated, but with a large sample size like 1000, the shape becomes close to uniform.

This is normal. Random data will never look perfectly flat, but it should look approximately flat.

Overlay PDFs of Normal, Exponential, and Uniform Distributions
Brief theory

Since Normal, Exponential, and Uniform distributions are continuous distributions, we compare them using their PDFs.

A PDF, or Probability Density Function, shows where values are more likely to occur. The total area under each PDF curve is equal to 1.

For this comparison, we use simple theoretical parameters:

Normal distribution: μ = 50, σ = 10

Exponential distribution: λ = 0.05

Uniform distribution: a = 20, b = 80


| Distribution | Shape        | Main idea                                     |
| ------------ | ------------ | --------------------------------------------- |
| Normal       | Bell-shaped  | Most values are near the mean                 |
| Exponential  | Right-skewed | Many small values, fewer large values         |
| Uniform      | Flat         | All values in the interval are equally likely |


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform

# -----------------------------
# Define x-axis range
# -----------------------------
x = np.linspace(0, 100, 1000)

# -----------------------------
# Normal Distribution parameters
# -----------------------------
mu = 50
sigma = 10

normal_pdf = norm.pdf(x, loc=mu, scale=sigma)

# -----------------------------
# Exponential Distribution parameters
# -----------------------------
lambda_exp = 0.05
scale_exp = 1 / lambda_exp   # scale = mean = 20

exponential_pdf = expon.pdf(x, loc=0, scale=scale_exp)

# -----------------------------
# Uniform Distribution parameters
# -----------------------------
a = 20
b = 80

uniform_pdf = uniform.pdf(x, loc=a, scale=b-a)

# -----------------------------
# Plot all PDFs together
# -----------------------------
plt.figure(figsize=(10, 6))

plt.plot(x, normal_pdf, linewidth=2, label="Normal PDF")
plt.plot(x, exponential_pdf, linewidth=2, label="Exponential PDF")
plt.plot(x, uniform_pdf, linewidth=2, label="Uniform PDF")

# Add mean lines
plt.axvline(mu, linestyle="--", alpha=0.6, label="Normal Mean")
plt.axvline(scale_exp, linestyle="--", alpha=0.6, label="Exponential Mean")
plt.axvline((a + b) / 2, linestyle="--", alpha=0.6, label="Uniform Mean")

plt.title("Overlay of Normal, Exponential, and Uniform PDFs")
plt.xlabel("x values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

Explanation

The overlay plot shows clear differences between the three continuous probability distributions.

The Normal distribution has a bell-shaped curve centered around its mean. Most of the probability density is concentrated near the center, and the curve decreases symmetrically on both sides.

The Exponential distribution is strongly right-skewed. It has the highest density near zero and gradually decreases as x increases. This means smaller values are more likely than larger values. This distribution is useful for modelling waiting times, such as the time until the next event occurs.

The Uniform distribution is flat between its lower and upper limits. This means every value between 20 and 80 has the same probability density. Unlike the normal distribution, it does not have a peak around the mean.

Overall, the plot helps compare shape, spread, and central tendency. The normal distribution is centered and symmetric, the exponential distribution is skewed to the right, and the uniform distribution gives equal likelihood to all values within its interval.

# 1. Introduction to the Binomial Distribution

The Binomial Distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials.

A trial can only have two outcomes:

*   Success
*   Failure

Examples:

*   Pass / Fail
*   Defective / Non-defective
*   Customer buys / does not buy
*   Heads / Tails

# 2. Bernoulli Trials

A Bernoulli trial is a single experiment with only two possible outcomes.

Example:



*   Tossing a coin once
*   Inspecting one product
*   Checking whether a customer clicked an ad


If:



*   Success probability = **p**
*   Failure probability = **1−p**


then each experiment is called a Bernoulli trial.

A Binomial experiment is simply:

*Repeating Bernoulli trials multiple times independently.*

# 3. Conditions for Binomial Distribution

A random variable follows a Binomial Distribution if:



1.   There is a fixed number of trials n
2.   Each trial has only two outcomes


1.   Trials are independent
2.   Probability of success p remains constant









```python
import numpy as np
import pandas as pd

# Reproducibility
np.random.seed(42)

# Parameters
n = 20          # number of trials
p = 0.3         # probability of success
size = 1000     # number of samples

# Generate binomial samples
samples = np.random.binomial(n=n, p=p, size=size)

# Create dataframe
df = pd.DataFrame({
    "Defective_Items": samples
})

# Display first rows
print(df.head())

# Save dataset
df.to_csv("binomial_dataset.csv", index=False)

print("\nDataset saved successfully.")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

# PMF values
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

# Plot
plt.figure(figsize=(10,5))

plt.bar(x, pmf)

plt.title("Binomial Distribution PMF")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")

plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Mean
mean = np.mean(samples)

# Variance
variance = np.var(samples)

print("Mean:", mean)
print("Variance:", variance)
```

# Binomial CDF

Yes — for Binomial distribution, the CDF (Cumulative Distribution Function) graph is very important because it shows:

The probability of getting less than or equal to k successes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 20
p = 0.3

# X values
x = np.arange(0, n + 1)

# CDF values
cdf = binom.cdf(x, n, p)

# Plot
plt.figure(figsize=(10,5))

plt.plot(x, cdf, marker='o')

plt.title("Binomial Distribution CDF")
plt.xlabel("Number of Successes")
plt.ylabel("Cumulative Probability")

plt.grid(True, alpha=0.3)
plt.show()
```
